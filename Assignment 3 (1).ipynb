```python

def number_value(a):
    try:
        if a > 0:
            print(a, "is positive")
        elif a==0:
            print (a, "is zero")
        else:
            print(a,"is negative ")
    except Exception as e:
        print(e)
```

```python
number_value(5)
```

```python
number_value(0)
```

```python
number_value(-5)
```

```python
number_value("svgvdj")
```

```python
def leap_year(a):
    try:
        if (a%4==0):
            if(a%100!=0):
                if(a%400!=0):  
                    print (a,"is leap year")
                else:
                    print(a,"not leap year")
            else:
                print(a,"not leap year")
        else:
                print(a,"not leap year")
    except Exception as e:
        print(e)
```

```python
leap_year(2000)
```

```python
leap_year(2004)
```

```python
4.	Write a Python Program to Check Prime Number?
```

```python
def check_prime(a):
    if a > 1:
        for i in range(2,a):
            if a%i==0:
                print(a, "is not prime")
                break
        else:
            print(a, "is prime")
    else:
        print("not prime")
        
```

```python
check_prime(3)
```

```python
check_prime(66)
```
